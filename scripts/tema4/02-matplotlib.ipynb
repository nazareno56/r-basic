# Mltiples grficos en una misma figura

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5.0, 0.1) # Vector de datos de 0 a 5 con saltos de 0.1
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1) # Defino una figura

plt.subplot(2,1,1) # 2 filas , 1 columna de graficos. Tercer parametro indica que es el primer grafico
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k') # Por un lado pinto los puntos y por el otro pinto la linea

plt.subplot(2,1,2)
plt.plot(x2, f(x2), 'g--')

plt.show()
```

```python
# Primer figura
plt.figure(1)
# Primer grafico
plt.subplot(2,1,1)
plt.plot([1,2,3])
# Segundo grafico
plt.subplot(2,1,2)
plt.plot([4,5,6])

# Segunda figura
plt.figure(2)
plt.plot([4,5,6]) # si no indico un subplot lo abarca todo

plt.figure(1)
plt.subplot(2,1,1)
plt.title("Esto es el primer ttulo")
```

# Textos en los grficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000) # randn(10000) 10000 datos en distribucion normal (campana de gauss)
```

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6) # histograma de frecuencia
# normed = 1 datos normalizados

plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$") # r"..." indica texto en raw de como que uno puede intruducir
plt.text(120, 0.015, r'$\mu = 100,\ \sigma=20$') # formato latex
plt.axis([20,180,0, 0.025]) # Limites de los ejes
plt.grid(True)


plt.show()
```

# Cambio de escala

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5 # media
sd = 0.3 # desviacion tipica
y = mu + sd*np.random.randn(1000) # distribucion normal
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logartmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthreshy=0.01) # linthreshy: cada cuanto quiero las lineas devisoras logaritmicas
plt.title("Escala Log Simtrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit") # la informacion se comprime en el centro y alarga en las puntas
plt.title("Escala logstica")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) # getCurrenAxis yaxis y le anulo el formato
plt.grid(True)

# ajustar los graficos dentro de la figure
plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35) 
# hspace: horizontal/high, wSpace: vertical/width

plt.show()
```

# Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) # creo 256 digitos entre -pi y pi
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure(figsize=(10,8))

plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "seno") # label permite dar una leyenda que sera
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno") # utilizada por plt.legend para representarlos
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)
# elijo los numeros a visualizar en los ejes
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])
plt.yticks(np.linspace(-1,1, 2, endpoint=True),
          ['-1',  '+1'])

# Muevo los ejes
ax = plt.gca() # get current axis
ax.spines['right'].set_color('none') # hago invicible al eje/marco derecho
ax.spines['top'].set_color('none') # hago invicible al eje/marco de arriba
ax.xaxis.set_ticks_position('bottom') # seteo un eje de las x/abcisas abajo
ax.spines['bottom'].set_position(('data', 0)) # y ahora lo puedo posicionar el medio en 0
ax.yaxis.set_ticks_position('left') # seteo un eje de las y/ordenadas a la izquierda
ax.spines['left'].set_position(('data', 0))# y ahora lo puedo posicionar al medio en 0

plt.legend(loc="upper left") # aado una leyenda arriba a la izquierda con los labels existentes

# Marcas y lineas sobre el grafico
x0 = 2*np.pi/3
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--") # linea
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue") # punto
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data", # donde va la anotacion y para que las coord xy coincidan con los datos
            xytext = (+20,+40), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2")) # anotacion

plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--") # linea
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green") # punto
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2")) # anotacion

#
for label in ax.get_xticklabels() + ax.get_yticklabels(): 
    # obtengo el conjunto de todas las etiquetas del eje de las x y eje de las y, las guardo en label.
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='grey', edgecolor='None', alpha = 0.2))

plt.show()
```
